```python
import pandas as pd
```

```python
data =  {
      'Patient_ID': [101,102,103,104],
      'Name': ['Alice', 'Bob', 'Cathy', 'Dan'],
    'Age': [29, 45, 31, 50],
    'Gender': ['F', 'M', 'F', 'M'],
    'Diagnosis': ['Diabetes', 'Hypertension', 'Diabetes', 'Heart Disease'],
    'Visit_Date': ['2024-12-01', '2024-12-05', '2025-01-10', '2025-01-15']
}

df = pd.DataFrame(data)
```
