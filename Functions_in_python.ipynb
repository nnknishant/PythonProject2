```python
import pandas as pd
```

```python
# Create a list in pthon

name = ["meena", "reena", "rohit", "juli", "meeenakshi", "raveena", "heena", "preeti"]
age = [25,24,25,26,26,24,15,15]
state = ["Bihar", "Jharkhand", "Bihar", "Delhi", "UP", "MP", "UP", "MP"]
city = ["Gaya", "Ranchi", "Gaya","Delhi", "Varanasi", "Bhopal", "varanasi", "Bhopal"]
salary = [25000,26000,24000,30000,28000,15000,16000,29000]

```

```python
# Merged the list in tupple

merged_list2  = list(zip(name, age, state,city,salary))
```

```python
print(merged_list2)
```

```python
# Now we create DataFrame from given avove list

df = pd.DataFrame(merged_list2, columns = ("name", "age", "state", "city", "salary"))
```
